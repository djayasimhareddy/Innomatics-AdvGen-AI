# Smart Systems Function-Based Problems

```python
# Problem 1: Smart Parking Lot Management System
def parking_system(capacity, logs):
    parked = 0
    peak = 0
    for log in logs:
        if log == 'IN':
            parked += 1
        elif log == 'OUT' and parked > 0:
            parked -= 1
        if parked > peak:
            peak = parked
    if parked > capacity:
        status = 'Full'
    else:
        status = 'Available'
    print('Currently Parked Vehicles:', parked)
    print('Parking Status:', status)
parking_system(50,['IN','IN','IN','OUT','IN','IN','OUT'])
```

```python
# Problem 2: Online Food Delivery Time Estimator
def delivery_time(distance, traffic, weather):
    time = distance * 5
    if traffic == 'High':
        time += 15
    if weather == 'Rainy':
        time += 10
    print('Estimated Delivery Time:', time, 'minutes')
delivery_time(8,'High','Rainy')
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer
def theatre_occupancy(total_seats, booked):
    count = len(booked)
    percentage = (count / total_seats) * 100
    if percentage == 100:
        status = 'Housefull'
    elif percentage >= 75:
        status = 'Almost Full'
    else:
        status = 'Seats Available'
    print('Occupancy:', int(percentage), '%')
    print('Show Status:', status)
theatre_occupancy(200,[1]*150)
```

```python
# Problem 4: Cloud Server Load Classification System
def server_load(cpu_readings):
    total = 0
    for c in cpu_readings:
        total += c
    avg = total / len(cpu_readings)
    if avg < 50:
        status = 'Normal'
    elif avg <= 80:
        status = 'Warning'
    else:
        status = 'Critical'
    print('Average CPU Load:', int(avg), '%')
    print('Server Status:', status)
server_load([45,60,70,85,90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor
def resource_usage(data):
    overused = []
    for item in data:
        if data[item] > 8:
            overused.append(item)
    if len(overused) > 0:
        alert = 'Yes'
    else:
        alert = 'No'
    print('Overused Resources:', ', '.join(overused))
    print('Energy Alert:', alert)
resource_usage({'Projector':6,'AC':9,'Lights':4})
```

```python
# Problem 6: Online Event Registration Capacity Controller
def event_registration(capacity, registrations):
    confirmed = 0
    waitlist = 0
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1
    print('Confirmed Registrations:', confirmed)
    print('Waitlisted Users:', waitlist)
    print('Registration Status: Closed')
event_registration(100,105)
```
