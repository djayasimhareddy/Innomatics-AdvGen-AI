```python
# Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        amount = int(input('Enter recharge amount: '))
        if amount < 50:
            print('Recharge amount must be at least ₹50')
        elif amount not in valid_plans:
            print('Invalid plan. Available plans:', valid_plans)
        else:
            print('Recharge Successful!')
            break
validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System
def check_inventory(products):
    for item in products:
        if products[item] < 15:
            print(item, '- Reorder Alert')
        else:
            print(item, '- Stock OK')
products = {'Rice': 10, 'Oil': 20, 'Sugar': 12}
check_inventory(products)
```

```python
# Problem 3: Student Result Processing System
def check_result(marks):
    total = 0
    for m in marks:
        total += m
    average = total / len(marks)
    if average >= 50:
        return 'Pass'
    else:
        return 'Fail'
marks = [60, 55, 40, 70]
print(check_result(marks))
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare():
    base_fare = 50
    rate_per_km = 12
    while True:
        km = float(input('Enter distance in km: '))
        peak = input('Is it peak hour? (yes/no): ')
        fare = base_fare + (km * rate_per_km)
        if peak.lower() == 'yes':
            fare = fare + (0.25 * fare)
        print('Total Fare: ₹', fare)
        break
calculate_fare()
```

```python
# Problem 5: Employee Attendance Eligibility Checker
def check_attendance(attendance):
    present = 0
    for day in attendance:
        if day == 'P':
            present += 1
    percentage = (present / len(attendance)) * 100
    if percentage >= 75:
        return 'Eligible'
    else:
        return 'Not Eligible'
attendance = ['P', 'P', 'A', 'P', 'P']
print(check_attendance(attendance))
```

```python
# Problem 6: Password Strength Checker
def check_password(password):
    has_digit = False
    has_special = False
    if len(password) < 8:
        return 'Weak Password'
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in '@#$':
            has_special = True
    if has_digit and has_special:
        return 'Strong Password'
    else:
        return 'Weak Password'
print(check_password('Test@123'))
```
