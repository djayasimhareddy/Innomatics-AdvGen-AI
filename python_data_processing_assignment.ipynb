```python
#question 1
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest = 0

for name in employees:
    if employees[name] > highest:
        highest = employees[name]

top_list = []

for name in employees:
    if employees[name] == highest:
        top_list.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_list), "(Score:", highest, ")")
```

```python
#question 2
query = "Buy mobile phone buy phone online"

query = query.lower()

words = query.split()

count = {}

for word in words:
    if word in count:
        count[word] += 1
    else:
        count[word] = 1

result = {}

for word in count:
    if count[word] > 1:
        result[word] = count[word]

print(result)
```

```python
#question 3
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

```python
#question 4
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    parts = email.split("@")
    domain = parts[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total = len(emails)

for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain + ":", int(percentage), "%")
```

```python
#question 5
sales = [1200, 1500, 900, 2200, 1400, 3000]

total = 0

for value in sales:
    total += value

average = total / len(sales)

limit = average + (0.3 * average)

for i in range(len(sales)):
    if sales[i] > limit:
        print("Day", i + 1, ":", sales[i])
```
