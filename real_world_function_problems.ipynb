# Real-World Function-Based Problem Solving

```python
# Problem 1: Social Media – Post Engagement Analyzer
def post_engagement(likes):
    total = 0
    for l in likes:
        total += l
    if total >= 1000:
        status = 'Viral Post'
    else:
        status = 'Normal Engagement'
    print('Total Likes:', total)
    print('Post Status:', status)
post_engagement([200,300,250,300])
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System
def medicine_stock(stock):
    if stock < 10:
        status = 'Low Stock Alert'
    else:
        status = 'Stock Sufficient'
    print('Medicine Stock:', stock)
    print('Status:', status)
medicine_stock(6)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_check(data, required):
    total = 0
    for r in data:
        total += r
    avg = total / len(data)
    if avg >= required:
        status = 'Adequate Rainfall'
    else:
        status = 'Inadequate Rainfall'
    print('Average Rainfall:', int(avg))
    print('Rainfall Status:', status)
rainfall_check([60,80,70,78],65)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def duplicate_accounts(users):
    unique = set(users)
    if len(unique) != len(users):
        print('Duplicate Accounts Found: Yes')
    else:
        print('Duplicate Accounts Found: No')
duplicate_accounts(['user1','user2','user1','user3'])
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker
def appointment(age):
    if age >= 18:
        status = 'Eligible'
    else:
        status = 'Not Eligible'
    print('Patient Age:', age)
    print('Eligibility Status:', status)
appointment(21)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def premium_crops(prices):
    premium = []
    for p in prices:
        if p > 2000:
            premium.append(p)
    print('Premium Crops:', premium)
premium_crops([1500,2500,1800,3200])
```

```python
# Problem 7: System Monitoring – Application Health Checker
def system_health(errors):
    if errors == 0:
        status = 'Healthy'
    elif errors <= 5:
        status = 'Minor Issues'
    else:
        status = 'Critical Issues'
    print('Error Count:', errors)
    print('System Status:', status)
system_health(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def transaction(amount):
    limit = 50000
    if amount <= limit:
        status = 'Approved'
    else:
        status = 'Rejected'
    print('Transaction Amount:', amount)
    print('Transaction Status:', status)
transaction(60000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System
def attendance_check(days):
    present = 0
    for d in days:
        if d == 'P':
            present += 1
    percentage = (present / len(days)) * 100
    if percentage >= 75:
        status = 'Eligible'
    else:
        status = 'Not Eligible'
    print('Attendance Percentage:', percentage)
    print('Exam Eligibility:', status)
attendance_check(['P','P','A','P','P'])
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    if bill < 500:
        status = 'Low Usage'
    elif bill <= 1500:
        status = 'Moderate Usage'
    else:
        status = 'High Usage'
    print('Total Bill:', bill)
    print('Usage Status:', status)
electricity_bill(250)
```
